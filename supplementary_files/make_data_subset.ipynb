```python
from Bio import SeqIO

# Specify the path to your FASTA file
fasta_file = "sequences.fasta"

# Open the FASTA file and count sequences
total_sequences = 0
with open(fasta_file, "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        total_sequences += 1

print("Total sequences:", total_sequences)
```

```python
sample_indices = np.random.choice(kmers.index, size=10000, replace=False)

# Sample rows using the generated indices
sample_df1 = kmers.loc[sample_indices]
sample_df2 = metadata.loc[sample_indices]
```

```python
import os
from glob import glob
import shutil

# List of files (replace with your list of file paths)
files = glob.glob()

# Sort the list of files
sorted_files = sorted(files)

# Destination folder
destination_folder = 'destination_folder'

for index in sample_indices:
    original_folder = sorted_files[index]

    # Copy the folder to the destination folder
    destination_path = os.path.join(destination_folder, os.path.basename(original_folder))
    shutil.copytree(original_folder, destination_path)
```
